# 02 — Preprocessing

This notebook cleans raw data, handles missing values/outliers, encodes variables, creates derived features, and saves a consistent analysis‑ready dataset.

```python
import os
import sys
import pandas as pd
import numpy as np

sys.path.append(os.path.abspath(".."))

from pathlib import Path

from src.data_prep import (
    load_raw_data, get_numeric_vars,
    get_categorical_vars, basic_cleaning, flag_outliers
)
from src.config import TARGET

# Paths
os.makedirs("../data/processed", exist_ok=True)
```

```python
# Load raw data
df = load_raw_data("../data/raw/migraine.csv")
```

```python
# Cleaned dataset
df_clean = basic_cleaning(df)
df_clean.head()
```

```python
# Outlier handling
NUMERIC_VARS = get_numeric_vars(df)

# Flag outliers instead of dropping
df_clean = df.copy()
for col in NUMERIC_VARS:
    df_clean = flag_outliers(df_clean, col)
```

The outliers detected using the IQR method have been reviewed and all of them are clinically coherent, so none have been removed. A flag has been created solely for traceability.
